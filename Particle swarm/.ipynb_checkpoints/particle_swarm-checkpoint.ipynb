```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def f(x, y):
    return abs(np.sin(x) * np.cos(y) * np.exp(abs(1 - (x**2 + y**2)**0.5 / np.pi)))
def g(x, y):
    return x * np.sin(np.pi * np.cos(x) * np.tan(y)) * (np.sin(y/x) / (1 + np.cos(y/x))) 
```

creating 2000 x and y between -10, 10 and computing the f function over them

```python
x, y = np.array(np.meshgrid(np.linspace(-10,10,2000), np.linspace(-10,10,2000)))
z = f(x, y)
z
```

ravel merges the x elements in one simple array form and we use the argmax to find the index of maximum elemnt. same to y...

```python
x_max = x.ravel()[z.argmax()]
y_max = y.ravel()[z.argmax()]
```

random initials (randn will generate negative elements also )

```python
# Create particles
n_particles = 20
np.random.seed(100)
X = np.random.rand(2, n_particles) * 20 - 10
V = np.random.randn(2, n_particles) * 0.1
X, V
```

making objects for calculating f on each random particle (personal and global bests...)

```python
# Initialize data
pbest = X
pbest_obj = f(X[0], X[1])
gbest = pbest[:, pbest_obj.argmax()]
gbest_obj = pbest_obj.max()
```

Drawing initialize

```python
fig, ax = plt.subplots(figsize=(8,6))
fig.set_tight_layout(True)
img = ax.imshow(z, extent=[-10, 10, -10, 10], origin='lower', cmap='viridis', alpha=0.5)
fig.colorbar(img, ax=ax)
ax.plot([x_max], [y_max], marker='x', markersize=10, color="red")
contours = ax.contour(x, y, z, 20, colors='black', alpha=0.4)
ax.clabel(contours, inline=True, fontsize=8, fmt="%.0f")
pbest_plot = ax.scatter(pbest[0], pbest[1], marker='o', color='blue', alpha=0.5)
p_plot = ax.scatter(X[0], X[1], marker='o', color='blue', alpha=0.5)
p_arrow = ax.quiver(X[0], X[1], V[0], V[1], color='blue', width=0.005, angles='xy', scale_units='xy', scale=1)
gbest_plot = plt.scatter([gbest[0]], [gbest[1]], marker='*', s=100, color='black', alpha=0.4)
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
```

Animation!

```python
def animate(i):
    title = 'Iteration {:02d}'.format(i)
    # Update params
    update()
    # Set picture
    ax.set_title(title)
    pbest_plot.set_offsets(pbest.T)
    p_plot.set_offsets(X.T)
    p_arrow.set_offsets(X.T)
    p_arrow.set_UVC(V[0], V[1])
    gbest_plot.set_offsets(gbest.reshape(1,-1))
    return ax, pbest_plot, p_plot, p_arrow, gbest_plot
```

Iterations Of Main Function

(instead of writing loops numpy operations do that in one line for every elements of that object)
(every numpy arrays have two rwos like V and X so for turning gbest in two rows we use reshape(-1,1) that means first element of table is unknown and it shoidl have 1 element in each row)

```python
def update():
    global V, X, pbest, pbest_obj, gbest, gbest_obj, c1, c2, w
    # Update params
    r1, r2 = np.random.rand(2)
    V = w * V + c1*r1*(pbest - X) + c2*r2*(gbest.reshape(-1,1)-X)
    X = X + V
    #print("XXXX", X)
    #print("VVVV", V)
    X[X < -10] = 10
    X[X > 10] = -10
    V[V < -1] = -1
    V[V > 1] = 1
    
    obj = f(X[0], X[1])
    pbest[:, (pbest_obj <= obj)] = X[:, (pbest_obj <= obj)]
    pbest_obj = np.array([pbest_obj, obj]).max(axis=0)
    gbest = pbest[:, pbest_obj.argmax()]
    gbest_obj = pbest_obj.max()
    print(gbest_obj)
```

```python
anim = FuncAnimation(fig, animate, frames=list(range(1,20)), interval=500, blit=False, repeat=True)
anim.save("PSO.gif", dpi=120, writer="imagemagick")
 
print("PSO found best solution at f({})={}".format(gbest, gbest_obj))
print("Global optimal at f({})={}".format([x_max,y_max], f(x_max,y_max)))
```

```python
import random, time
def init(n_particles, a, b, c):
    # Create particles
    #n_particles = 20
    global V, X, pbest, pbest_obj, gbest, gbest_obj, c1, c2, w
    c1 = a
    c2 = b
    w = c
    np.random.seed(100)
    X = np.random.rand(2, n_particles) * 200 - 100
    V = np.random.rand(2, n_particles) * 0.2 - 0.1

    pbest = X
    pbest_obj = g(X[0], X[1])
    gbest = pbest[:, pbest_obj.argmin()]
    gbest_obj = pbest_obj.min()
    
    
def update():
    global V, X, pbest, pbest_obj, gbest, gbest_obj, c1, c2, w
    # Update params
    r1, r2 = np.random.rand(2)
    V = w * V + c1*r1*(pbest - X) + c2*r2*(gbest.reshape(-1,1)-X)
    X = X + V
    #print("XXXX", X)
    #print("VVVV", V)
    X[X < -100] = random.random() * 100
    X[X > 100] =  random.random() * -100
    V[V < -1] = -1
    V[V > 1] = 1
    
    obj = g(X[0], X[1])
    pbest[:, (pbest_obj >= obj)] = X[:, (pbest_obj >= obj)]
    pbest_obj = np.array([pbest_obj, obj]).min(axis=0)
    gbest = pbest[:, pbest_obj.argmin()]
    gbest_obj = pbest_obj.min()
    print(gbest_obj, end = '\r', flush=True)
init(400, 0.005, 0.005, 0.55)
t1 = time.time()
for i in range(10000):
    update()
print("PSO found best solution at f({})={}".format(gbest, gbest_obj))
print("Time: ", time.time() - t1)
```
